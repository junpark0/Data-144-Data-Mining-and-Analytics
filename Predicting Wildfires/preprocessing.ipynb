```python
fire1 = pd.read_csv("fire.csv")
```

```python
fire2 = fire1.drop(columns = ["AdminUnit", "CanonicalUrl", "ConditionStatement", "ControlStatement", "SearchKeywords", 
                            "Status", "UniqueId", "Location", "PercentContained", "SearchDescription", "Active", 
                           "CountyIds", "Featured", "Final", "Public", "Updated", "Name", "FuelType"])
```

```python
fire2 = fire2.fillna(0)
```

```python
#deleted rows where extinguished date was unavailable
fire3 = fire2[fire2["Extinguished"] != 0]
```

```python
#changed date format for extinguished fires
pattern = '\d{4}-\d{2}-\d{2}'
dates = list()
for x in fire3["Extinguished"]:
    match = re.search(pattern, x)
    date = match.group()
    dates.append(date)
fire3["Extinguished"] = dates
fire3
```

```python
#changed date format for fire start date
sdates = list()
for x in fire3["Started"]:
    match = re.search(pattern, x)
    date = match.group()
    sdates.append(date)
fire3["Started"] = sdates
fire3
```

```python
edate_obj = []
for x in fire3["Extinguished"]:
    date_time_obj = datetime.strptime(x, '%Y-%m-%d')
    edate_obj.append(date_time_obj)

fire3["Extinguished"] = edate_obj
```

```python
sdate_obj = []
for x in fire3["Started"]:
    date_time_obj = datetime.strptime(x, '%Y-%m-%d')
    sdate_obj.append(date_time_obj)

fire3["Started"] = sdate_obj
```

```python
#new feature for fire duration
fire3["Days Elapsed"] = fire3["Extinguished"] - fire3["Started"]
fire3.head()
```

```python
#got rid of extraneous data where extuinguish date was before start date
fire4 = fire3[fire3["Days Elapsed"] >= "0 Days"]
```

```python
fire4["Latitude"] = fire4["Latitude"].round(3)
fire4["Longitude"] = fire4["Longitude"].round(3)
```

```python
len(fire4)
```

```python
weather = pd.read_csv("weather.csv")
weather.head()
```

```python
fire4.head()
```

```python
date_obj = []
for x in weather["date"]:
    date_time_obj = datetime.strptime(x, '%Y-%m-%d')
    date_obj.append(date_time_obj)

weather["date"] = date_obj
```

```python
#added in weather data 
fire5 = pd.merge(fire4, weather, left_on = ['Latitude', 'Longitude', 'Started'], 
                 right_on = ['latitude', 'longitude', 'date'])
fire5.head()
```

```python
fire5["Average Temp F"] = (fire5["maxtempF"] + fire5["mintempF"]) / 2
```

```python
fire6 = fire5[["Latitude", "Longitude", "Started", "Extinguished", "Days Elapsed", "AcresBurned", "MajorIncident",
               "Average Temp F", "uvIndex", "sunHour", "totalSnow_cm", "AirTankers", "CalFireIncident", "Counties", 
               "CrewsInvolved", "Dozers", "Engines", "Fatalities", "Helicopters", "Injuries", "PersonnelInvolved",
               "StructuresDamaged", "StructuresDestroyed", "StructuresEvacuated", "StructuresThreatened", 
               "WaterTenders"]]
```

```python
fire6["Days Elapsed"] = fire6["Days Elapsed"].dt.days
```

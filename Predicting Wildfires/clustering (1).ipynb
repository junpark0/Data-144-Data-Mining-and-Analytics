```python
def bestK(df, features):
    i = 2
    k = i
    score_to_beat = 0
    current_score = 0
    X = np.array(df[features].dropna())

    while i <= 8:
      kmeans = KMeans(n_clusters=i, random_state=42).fit(X)
      #df.dropna()
      current_score = silhouette_score(X, kmeans.labels_)
      if (abs(current_score-1) < abs(score_to_beat-1)):
        k = i
        score_to_beat = current_score
      i = i + 1
    return k
    raise NotImplementedError()

bestK = bestK(fire6, ['Average Temp F', 'uvIndex', 'sunHour'])
print("Best k with : ['Average Temp F', 'uvIndex', 'sunHour']: ", bestK)
```

```python
def dataPointsPerCluster(df, features):
    k = bestK
    X = np.array(df[features].dropna())
    kmeans = KMeans(n_clusters=k).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    kmeans_arr = kmeans.labels_
    Dict = {}
    c = 0
    while c < k:
      Dict[c] = 0
      for j in kmeans_arr:
        if j == c:
          Dict[c] = Dict[c] + 1
      c = c + 1
    return Dict
    raise NotImplementedError()

print(dataPointsPerCluster(fire6, ['Average Temp F', 'uvIndex', 'sunHour']))
```

```python
def clusterLabelsArr(df, features):
    k= bestK
    X = np.array(df[features].dropna())
    kmeans = KMeans(n_clusters=k).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    kmeans_arr = kmeans.labels_
    #return cluster
    return kmeans_arr

labels = clusterLabelsArr(fire6, ['Average Temp F', 'uvIndex', 'sunHour'])
print(labels)
```

```python
def clusterCentroid(df, features):
    k= bestK
    X = np.array(df[features].dropna())
    kmeans = KMeans(n_clusters=k).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    kmeans_arr = kmeans.labels_
    cluster = kmeans.cluster_centers_
    #return cluster
    return cluster

print(clusterCentroid(fire6, ['Average Temp F', 'uvIndex', 'sunHour']))
```

```python
def averageFeature(df, features, feature):
    q_arr = df[features].dropna()
    q3_df = df[features].dropna()[[feature]]
    kmeans = KMeans(n_clusters=bestK).fit(np.array(q_arr.drop(columns=[feature])))
    q3_df['clusters'] = kmeans.labels_
    cluster_arr_avg = np.array(q3_df.groupby('clusters').mean()[feature])
    return cluster_arr_avg

print("Average Temp F per Cluster: ", averageFeature(fire6, ['Average Temp F', 'uvIndex', 'sunHour'], 'Average Temp F'))
print("Average uvIndex per Cluster: ", averageFeature(fire6, ['Average Temp F', 'uvIndex', 'sunHour'], 'uvIndex'))
print("Average sunHour per Cluster: ", averageFeature(fire6, ['Average Temp F', 'uvIndex', 'sunHour'], 'sunHour'))
```

```python
def silhoutteScore(df, features):
    X = np.array(df[features].dropna())
    kmeans = KMeans(n_clusters=bestK).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    return score

print("Silhoutte score: ", silhoutteScore(fire6, ['Average Temp F', 'uvIndex', 'sunHour']))
```

```python
Xc = df[["Latitude", "Longitude", "Average Temp F", "uvIndex", "sunHour", "totalSnow_cm"]]
yc = df[["MajorIncident"]]

Xr = df.drop(columns = ["Days Elapsed", "AcresBurned"])
yrd = df[["Days Elapsed"]]
yrb = df[["AcresBurned"]]
```

```python
#classification train test split 
Xc_tr, Xc_te, yc_tr, yc_te = train_test_split(Xc, yc, test_size=0.2, random_state=42)

#regression for days elapsed train test split
Xrd_tr, Xrd_te, yrd_tr, yrd_te = train_test_split(Xr, yrd, test_size=0.2, random_state=42)

#regression for acres burned train test split
Xrb_tr, Xrb_te, yrb_tr, yrb_te = train_test_split(Xr, yrb, test_size=0.2, random_state=42)
```

```python
import scipy.stats as stats
label = preprocessing.LabelEncoder()

Y_train = label.fit_transform(yrb_tr['AcresBurned'])

Xrb_tr['Major_Incident'] = Xrb_tr['MajorIncident']
Xrb_tr['Average_Temp_F'] = stats.zscore(Xrb_tr['Average Temp F'])
Xrb_tr['Counties'] = label.fit_transform(Xrb_tr['Counties'])

Xrb_te['Major_Incident'] = Xrb_te['MajorIncident']
Xrb_te['Average_Temp_F'] = stats.zscore(Xrb_te['Average Temp F'])
Xrb_te['Counties'] = label.fit_transform(Xrb_te['Counties'])


X_train_dum = Xrb_tr[['Major_Incident', 'Dozers', 'Helicopters', 'Injuries', 'StructuresDestroyed', 'StructuresThreatened']]
X_test_dum = Xrb_te[['Major_Incident','Dozers', 'Helicopters', 'Injuries', 'StructuresDestroyed', 'StructuresThreatened']]

clf = MLPClassifier(hidden_layer_sizes=(50,100,150), max_iter= 100,
                     solver='lbfgs', verbose=1,  random_state=43)  #sgd = Stochastic Gradient descent. Also, check 'lbfgs’, ‘adam’}
                                                                #Default activation is 'relu'

clf.fit(X_train_dum, Y_train)

y_pred_train=clf.predict(X_train_dum)
print('RMSE on training---')
y_pred_train=clf.predict(X_train_dum)
print(np.sqrt(mean_squared_error(Y_train,y_pred_train)))
```

```python
import scipy.stats as stats
label = preprocessing.LabelEncoder()

Y_train = label.fit_transform(yrd_tr['Days Elapsed'])

Xrd_tr['Major_Incident'] = Xrd_tr['MajorIncident']
Xrd_tr['Average_Temp_F'] = stats.zscore(Xrd_tr['Average Temp F'])
Xrd_tr['Counties'] = label.fit_transform(Xrd_tr['Counties'])


Xrd_te['Major_Incident'] = Xrd_te['MajorIncident']
Xrd_te['Average_Temp_F'] = stats.zscore(Xrd_te['Average Temp F'])
Xrd_te['Counties'] = label.fit_transform(Xrd_te['Counties'])



X_train_dum = Xrd_tr[['Major_Incident',  'Average_Temp_F', 'Counties', 'Latitude', 'Longitude',  'Helicopters', 'Injuries', 'StructuresDestroyed', 'StructuresThreatened']]
X_test_dum = Xrd_te[['Major_Incident',  'Average_Temp_F', 'Counties', 'Latitude', 'Longitude', 'Helicopters', 'Injuries', 'StructuresDestroyed', 'StructuresThreatened']]

clf = MLPClassifier(hidden_layer_sizes=(50,100,150), max_iter= 100,
                     solver='lbfgs', verbose=1,  random_state=43)  #sgd = Stochastic Gradient descent. Also, check 'lbfgs’, ‘adam’}
                                                                #Default activation is 'relu'

clf.fit(X_train_dum, Y_train)

y_pred_train=clf.predict(X_train_dum)
print('RMSE on training---')
y_pred_train=clf.predict(X_train_dum)
print(np.sqrt(mean_squared_error(Y_train,y_pred_train)))
```
